```python
df.drop("Town",inplace=True,axis=1)
```

```python
df.head()
x = df[["Area"]]
y= df["Price"]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(df["Area"],df["Price"],label="actual values")
plt.plot(df["Area"],y_pred,label="Predicted values")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```
