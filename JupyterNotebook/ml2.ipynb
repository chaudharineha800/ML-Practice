## Imbalanced dataset:
### A dataset where the target class has an uneven distribution of observation


#### Machine learnign algorithm try to maximise overall accuracy
#### if the model predicts only the majority class but the model failed to detect all minority cases htis is called the ACCURACY PARADOX

## Techniques of balancing

### 1.Undersampling:
#### Randomly delete rows from the Majority Class until it matches the size of the Minirity Class.
#### Pros: Very fast training because the dataset becomes much smaller
#### Cons:We discard a huge amount of useful informaion from the majority class.

### 2.OverSampling
#### Randomly duplicates rows from the Minority class until it matches the size of the Majority Class
#### Pros:No information from the dataset is lost
#### Cons:The model may memorize those same 100 faculty cars instead of learnign general patterns


### 3.SMOTE(Synthetic Minority Over-sampling)
#### SMOTE creates new synthetic samples insted of duplicating existing ones
#### It uses K-Nearest Neighbors to generate new data points between existing minirity samoles

```python
import pandas as pd
```

```python
data = {
    "Age":[22,45,63,23,44,56,23,78,46,23,56,23,45,89],
    "Salary":[20000,25000,50000,52000,48000,60000,22000,34000,40000,34000,27000,55000,78000,9090],
    "Purchased":[0,0,1,1,1,1,1,0,1,1,1,0,0,0]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.duplicated().sum()
```

```python
df["Purchased"].value_counts()
```

```python
x = df.drop("Purchased",axis=1)
y = df["Purchased"]
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
rv = RandomOverSampler()
```

```python
rvx,rvy=rv.fit_resample(x,y)
```

```python
rvy.value_counts()
```

```python
rvx.duplicated().sum()
```

```python
x = df.drop("Purchased",axis=1)
y = df["Purchased"] 
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rs = RandomUnderSampler()
```

```python
rsx,rsy = rs.fit_resample(x,y)
```

```python
rsy.value_counts()
```

```python
rsx.duplicated().sum()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
rx = SMOTE()
```

```python
rxx,rxy=rx.fit_resample(x,y)
```

```python
rxy.value_counts()
```
